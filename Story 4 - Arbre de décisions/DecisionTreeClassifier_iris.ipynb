```python
# Importation du jeu de données Iris et du module pour les arbres de décision
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib
```

```python
# Chargement du jeu de données Iris
iris = load_iris()  # Le jeu de données contient des caractéristiques de fleurs (ex : longueur des pétales) et leur classe (espèce)

# Séparation des données en variables explicatives (X) et cible (y)
X, y = iris.data, iris.target  # X contient les caractéristiques, y contient les étiquettes des classes (espèces)
```

```python
# Création d'un classifieur basé sur un arbre de décision
clf = tree.DecisionTreeClassifier()

# Entraînement du classifieur sur les données
clf = clf.fit(X, y)  # Le modèle est ajusté sur les données pour apprendre à prédire l'espèce des fleurs
```

```python
# Visualisation de l'arbre de décision formé
tree.plot_tree(clf, filled=True)  # Affiche une représentation graphique de l'arbre de décision entraîné
```
